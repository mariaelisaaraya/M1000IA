```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd
```

```python


# Ruta del archivo CSV
file_path = r"https://raw.githubusercontent.com/mariaelisaaraya/M1000IA/refs/heads/master/metadatos_actualizados_limpios_normalizados.csv"

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv(file_path)

# Mostrar las primeras 5 filas del DataFrame
df.head()
```

```python
# Mostrar el número de filas y columnas
df.shape
```

```python
# Mostrar los tipos de datos de cada columna
df.dtypes
```

```python
# Mostrar la cantidad de valores nulos por columna
df.isnull().sum()
```

```python
# Definir rangos de edad
bins = [0, 18, 30, 40, 50, 60, 70, 80, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '61-70', '71-80', '80+']
df['age_range'] = pd.cut(df['age'], bins=bins, labels=labels)

# Calcular la distribución de tipo de lesión por rango de edad
distribution_age_dx = df.groupby(['age_range', 'dx']).size().unstack(fill_value=0)

# Mostrar la tabla de distribución
print(distribution_age_dx)
```

```python
# Visualizar la distribución con un gráfico de barras apiladas
distribution_age_dx.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Distribución de Tipo de Lesión por Rango de Edad')
plt.xlabel('Rango de Edad')
plt.ylabel('Cantidad de Lesiones')
plt.legend(title='Tipo de Lesión')
plt.show()
```

```python
# Calcular la distribución de tipo de lesión por sexo
distribution_sex_dx = df.groupby(['sex', 'dx']).size().unstack(fill_value=0)

# Mostrar la tabla de distribución
print(distribution_sex_dx)
```

```python
# Visualizar la distribución con un gráfico de barras apiladas
distribution_sex_dx.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.title('Distribución de Tipo de Lesión por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Cantidad de Lesiones')
plt.legend(title='Tipo de Lesión')
plt.show()
```

```python
# Calcular la distribución para la clasificación benigna y maligna
distribution_classification = df.groupby(['classification', 'dx']).size().unstack(fill_value=0)

# Mostrar la tabla de distribución
print(distribution_classification)
```

```python
# Visualizar la distribución con un gráfico de barras apiladas
distribution_classification.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Distribución de Tipo de Lesión para Clasificación Benigna y Maligna')
plt.xlabel('Clasificación')
plt.ylabel('Cantidad de Lesiones')
plt.legend(title='Tipo de Lesión')
plt.show()
```

```python

# Agrupar por sexo, tipo de lesión (dx) y clasificación (maligno o benigno)
distribution = df.groupby(['sex', 'dx', 'classification']).size().reset_index(name='count')

# Crear la gráfica de barras apiladas con seaborn
plt.figure(figsize=(12, 8))
sns.barplot(data=distribution, x='dx', y='count', hue='classification', ci=None, palette='muted')

# Ajustar etiquetas y título
plt.title('Distribución de tipos de lesiones por sexo y clasificación (maligna o benigna)')
plt.ylabel('Cantidad de lesiones')
plt.xlabel('Tipo de lesión (dx)')
plt.xticks(rotation=45)

# Dividir la gráfica por sexo
g = sns.FacetGrid(distribution, col="sex", height=6, aspect=1.5)
g.map_dataframe(sns.barplot, x='dx', y='count', hue='classification', palette='muted')
g.add_legend()

# Mostrar la gráfica
plt.show()
```

```python

# Agrupar por sexo, tipo de lesión (dx) y clasificación (maligno o benigno)
distribution = df.groupby(['age_range', 'dx', 'classification']).size().reset_index(name='count')

# Crear la gráfica de barras apiladas con seaborn
plt.figure(figsize=(12, 8))
sns.barplot(data=distribution, x='age_range', y='count', hue='classification', ci=None, palette='muted')

# Ajustar etiquetas y título
plt.title('Distribución de tipos de lesiones por rango de edad y clasificación (maligna o benigna)')
plt.ylabel('Cantidad de lesiones')
plt.xlabel('Rango Edad (age_range)')
plt.xticks(rotation=45)

# Dividir la gráfica por sexo
g = sns.FacetGrid(distribution, col="age_range", height=6, aspect=1.5)
g.map_dataframe(sns.barplot, x='dx', y='count', hue='classification', palette='muted')
g.add_legend()

# Mostrar la gráfica
plt.show()
```
