```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file = "metadatos_actualizados_limpios.csv"

# Leer el archivo CSV desde la URL
df = pd.read_csv(file)
```

```python
# Pasar todos los datos a minúsculas
df = df.apply(lambda x: x.str.lower() if x.dtype == "object" else x)

df.head(-10)
```

```python
# Contar valores únicos por cada columna
print("\nValores únicos por columna:")
valores_unicos = df.nunique()
print(valores_unicos)
```

```python
# Cambiar el nombre de la columna "Clasificacion" a "classification" y "Dataset" a "dataset"
df.rename(columns={"Clasificacion": "classification", "Dataset": "dataset"}, inplace=True)

df.head(-10)
```

```python
# Guardar el dataframe en un archivo CSV
df.to_csv("metadatos_actualizados_limpios_normalizados.csv", index=False)
```
